```python
# Sejarah Python
def python_history():
    sejarah = '''
    Pencipta bahasa pemrograman Python yakni Guido Van Rossum terinspirasi dari acara Mothy Python Flying Circus.
    Pada 20 Februari 1991 pun akhirnya Van Rossum akhirnya mulai mengembangkan bahasa pemrograman Python. Tujuannya :
    - Bahasa pemrograman Python cocok digunakan sehari-hari, sebab dalam pengembangannya cukup singkat
    - Open source, bisa dikembangkan oleh siapa saja
    - Bahasa yang mudah dipahami dan intuitif sebanding dengan bahasa pemrograman lain
    - Menggunakan kode bahasa Inggris biasa
    '''
    return print(sejarah)

# Keunggulan python
def python_benefits():
    keuntungan = '''
    Kelebihan bahasa pemrograman Python:
    - Mudah digunakan untuk menulis di perangkat lunak baru
    - Mudah untuk diajarkan
    - Mudah dipahami
    - Mudah didapatkan, karena mudah diinstal. Selain itu, Python juga multiplatform
    '''
    return print(keuntungan)

#Fungsi Kalkulator
def calculator():
        # Fungsi penambahan
        def penambahan(x, y):
            return x + y

        # Fungsi pengurangan
        def pengurangan(x, y):
            return x - y

        # Fungsi perkalian
        def perkalian(x, y):
            return x * y

        # Fungsi pembagian
        def pembagian(x, y):
            return x / y


        print("Pilih Operasi Hitung.")
        print("1.Penambahan")
        print("2.Pengurangan")
        print("3.Perkalian")
        print("4.Pembagian")

        while True:
            # Masukan pilihan operasi hitung
            choice = input("Pilihan (1/2/3/4): ")

            # Pembuatan exeption
            if choice in ('1', '2', '3', '4'):
                try:
                    num1 = float(input("Masukan angka pertama: "))
                    num2 = float(input("Masukan angka kedua: "))

                    if choice == '1':
                        print(num1, "+", num2, "=", penambahan(num1, num2))

                    elif choice == '2':
                        print(num1, "-", num2, "=", pengurangan(num1, num2))

                    elif choice == '3':
                        print(num1, "*", num2, "=", perkalian(num1, num2))

                    elif choice == '4':
                        print(num1, "/", num2, "=", pembagian(num1, num2))
                    break
                except ValueError as ve:
                    print('Terjadi error pada nilai input!!!')
            else:
                print("Tolong masukan pilihan yang tepat")
                
#Print jawaban
print('Apa itu python dan bagaimana sejarahnya?') #Soal 1
python_history() #Jawaban 1

print('Apa saja keunggulan python?') #Soal 2
python_benefits() #Jawaban 2

print('Buatlah program sederhana kalkulator!') #Soal 3
calculator() #Jawaban 3
```
